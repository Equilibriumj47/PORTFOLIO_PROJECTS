## **Data Preparation**

### **Data Separation as x & y**

### **Checking for outliers for the Y column (MEDV)**

```python
# Isolating the Y column
y = df_clean['MEDV']
y
```

```python

# finding the 1st quantile and 3rd quantile of MEDV col
q1, q3 = np.percentile(y, [25, 75])
print('Q1: ', q1)
print('Q3: ', q3)
# Find the IQR
IQR = q3 -q1
print("The IQR value is: ", IQR)
# Find the lower band and higher band values

lower_band_val = q1 - (1.5 * IQR)
higher_band_val = q3 + (1.5 * IQR)

print("The Lower Band Value is: ", lower_band_val)
print("The Higher Band Value is: ", higher_band_val)

filt_1 = y < 5.06
filt_2 = y > 36.96

#Checking for outliers:
print('The Lower_outliers are: ')
Lower_outliers = y[filt_1]
print(Lower_outliers)
print('\n')
print('The Upper_outliers are: ')
Upper_outliers = y[filt_2]
print(Upper_outliers)
```

### Isolating the X Data

```python
x = df_clean.drop('MEDV', axis = 1)
x
```

## Data Spliting

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 100)
```

```python
x_train
```

```python
x_test
```

# Model Building

## Linear Regression

### Training the Model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x_train,y_train)
```

```python
print(lr.coef_)
```

### Applying the Model to Make Prediction

```python
y_lr_train_pred = lr.predict(x_train)

y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

### Evaluate Performance

```python
y_train
```

```python
y_lr_train_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
pred = ['Linear_Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]


pred_df = pd.DataFrame(pred).transpose()
pred_df.columns = ['Method','Train_MSE', 'Train_R2', 'Test_MSE', 'Test_R2']
```

### Random Forest

### Training Model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth = 2, random_state = 100)
rf.fit(x_train, y_train)
```

### Applying the Model to Make Prediction

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

### Evaluating Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_result = ['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse,rf_test_r2]
rf_result_df = pd.DataFrame(rf_result).transpose()
rf_result_df.columns = ['Method','Train_MSE', 'Train_R2', 'Test_MSE', 'Test_R2']

```

### Feature Selection

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
y = np.round(df_clean['MEDV'])
#Apply SelectKBest class to extract top 5 best features
bestfeatures = SelectKBest(score_func=chi2, k=5)
fit = bestfeatures.fit(x,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)
# Concat two dataframes for better visualization
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score'] #naming the dataframe columns
featureScores
```

```python
#print 5 best features
print(featureScores.nlargest(5,'Score'))
```
